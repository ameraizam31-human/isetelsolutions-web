```python


def divider():
    print("-" * 40)

def get_int_list(prompt):
    while True:
        try:
            raw = input(prompt)
            return list(map(int, raw.split()))
        except ValueError:
            print("  Invalid input. Try again.")


#Q1 -Count Positive and Negative using Loop

def q1_count_loop():
    print("\nQ1 - Count Positive & Negative (Loop)")
    numbers = get_int_list("Enter numbers : ")

    pos = 0
    neg = 0
    for n in numbers:
        if n > 0:
            pos += 1
        elif n < 0:
            neg += 1

    print("No. Of Positive :", pos)
    print("No. Of Negative :", neg)
    divider()


#Q2 - Count Positive and Negative using List Comprehension
def q2_count_comprehension():
    print("\nQ2 - Count Positive & Negative (Comprehension)")
    numbers = get_int_list("Enter numbers : ")

    pos = len([n for n in numbers if n > 0])
    neg = len([n for n in numbers if n < 0])

    print("No. Of Positive :", pos)
    print("No. Of Negative :", neg)
    divider()


#Q3 - Remove Multiple Elements
def q3_remove_elements():
    print("\nQ3 - Remove Multiple Elements")
    a = get_int_list("Enter main list   : ")
    b = get_int_list("Items to remove   : ")

    result = [x for x in a if x not in b]

    print("Original :", a)
    print("Removed  :", b)
    print("Result   :", result)
    divider()


# Q4 - Remove Duplicates

def q4_remove_duplicates():
    print("\nQ4 - Remove Duplicates")
    a = get_int_list("Enter list : ")

    seen = []
    for val in a:
        if val not in seen:
            seen.append(val)

    print("Original :", a)
    print("Result   :", seen)
    divider()


# Q5 - Sum of Dictionary Values

def q5_sum_dictionary():
    print("\nQ5 - Sum of Dictionary Values")

    d = {}
    n = int(input("How many keys? "))
    for _ in range(n):
        key = input("Key   : ")
        val = int(input("Value : "))
        d[key] = val

    total = sum(d.values())
    print("Dictionary :", d)
    print("Total Sum  :", total)
    divider()


# Q6 - Key with Maximum Unique Values

def q6_max_unique():
    print("\nQ6 - Key with Maximum Unique Values")

    d = {}
    n = int(input("How many keys? "))
    for _ in range(n):
        key = input("Key    : ")
        val = get_int_list("Values : ")
        d[key] = val

    best_key = None
    highest = 0

    for k, v in d.items():
        count = len(set(v))
        if count > highest:
            highest = count
            best_key = k

    print("Dictionary :", d)
    print("Key with most unique values :", best_key)
    divider()


# Q7 - Sales Aggregation

def q7_sales():
    print("\nQ7 - Sales Aggregation")

    sales = []
    n = int(input("How many records? "))
    for _ in range(n):
        product = input("Product  : ")
        qty = int(input("Quantity : "))
        sales.append((product, qty))

    total = {}
    for product, qty in sales:
        total[product] = total.get(product, 0) + qty

    print("Result :", total)
    divider()


# Q8 - Undo Feature using Stack

def q8_undo():
    print("\nQ8 - Undo Feature")
    print("Type action, 'delete' to undo, 'done' to finish")

    stack = []
    while True:
        action = input("Action : ").strip()
        if action == "done":
            break
        elif action == "delete":
            if stack:
                removed = stack.pop()
                print("Undone :", removed)
            else:
                print("Nothing to undo.")
        else:
            stack.append(action)

    print("Final :", stack)
    divider()


# Q9 - Two Sum

def q9_two_sum():
    print("\nQ9 - Two Sum")
    weights = get_int_list("Enter weights : ")
    target = int(input("Target value  : "))

    seen = {}
    print("Pairs that sum to", target, ":")
    for i, w in enumerate(weights):
        diff = target - w
        if diff in seen:
            print(" Index", seen[diff], "and", i)
        seen[w] = i
    divider()


# Q10 - Price Update using map and lambda

def q10_price_update():
    print("\nQ10 - Price Update +10%")
    prices = get_int_list("Enter prices : ")

    updated = list(map(lambda x: round(x * 1.10, 2), prices))
    print("Before :", prices)
    print("After  :", updated)
    divider()


# Q11 - Employee Bonus

def q11_bonus():
    print("\nQ11 - Employee Bonus")
    print("Rule : >= 50000 gets 20%, else 10%")
    salaries = get_int_list("Enter salaries : ")

    result = list(map(lambda s: s * 1.2 if s >= 50000 else s * 1.1, salaries))
    print("Before :", salaries)
    print("After  :", result)
    divider()


# Q12 - Order Filtering

def q12_filter_orders():
    print("\nQ12 - Order Filtering")
    print("Rule : amount > 1000 and status = completed")

    orders = []
    n = int(input("How many orders? "))
    for _ in range(n):
        amount = int(input("Amount : "))
        status = input("Status (completed/cancelled) : ").strip()
        orders.append({"amount": amount, "status": status})

    valid = list(filter(
        lambda o: o["amount"] > 1000 and o["status"] == "completed",
        orders
    ))

    print("All Orders   :", orders)
    print("Valid Orders :", valid)
    divider()


# Q13 - Data Cleaning Pipeline

def q13_pipeline():
    print("\nQ13 - Data Cleaning Pipeline")
    print("Step 1 : Remove negatives")
    print("Step 2 : Square the numbers")
    print("Step 3 : Keep only values > 50")

    data = get_int_list("Enter data : ")

    result = list(
        filter(lambda x: x > 50,
        map(lambda x: x**2,
        filter(lambda x: x > 0, data)))
    )

    print("Input  :", data)
    print("Output :", result)
    divider()


# Main Menu

def main_menu():
    options = {
        "1":  q1_count_loop,
        "2":  q2_count_comprehension,
        "3":  q3_remove_elements,
        "4":  q4_remove_duplicates,
        "5":  q5_sum_dictionary,
        "6":  q6_max_unique,
        "7":  q7_sales,
        "8":  q8_undo,
        "9":  q9_two_sum,
        "10": q10_price_update,
        "11": q11_bonus,
        "12": q12_filter_orders,
        "13": q13_pipeline,
        "0":  None
    }

    while True:
        choice = input("\nSelect question [0-13] : ").strip()

        if choice == "0":
            print("Goodbye!")
            break
        elif choice in options:
            options[choice]()
        else:
            print("Invalid. Enter 0 to 13.")


main_menu()
```
